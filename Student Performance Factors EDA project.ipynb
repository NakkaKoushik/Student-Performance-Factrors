### Data Cleaning

```python
# To remove columns
data.drop(['Family_Income', 'Parental_Education_Level', 'Peer_Influence', 'Distance_from_Home', 'Learning_Disabilities'],axis=1)
```

```python
#To check null values
data.isnull().sum()
```

```python
#To remove null values
data.dropna(axis=0,inplace=True)
```

```python
# Null values are removed
data.isnull().sum()
```

### Data Visualization

```python
# Plot histogram of a single feature
plt.figure(figsize=(16,6))
plt.hist(data['Attendance'], bins=30,  edgecolor='black')
plt.title('Histogram For Attendance')
plt.xlabel('Attendance')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot boxplot in a single feature
plt.figure(figsize=[16,6])
sns.boxplot(x=data['Attendance'])
plt.title('Boxplot for Attandance')
plt.xlabel('Attendance')
plt.show()
```

```python
# Basic Barplot
plt.figure(figsize=[16,6])
sns.barplot(x='Gender', y='Attendance', data=data)
plt.title('Barplot of data')
plt.xlabel('Gender')
plt.ylabel('Attendance')
plt.show()
```

```python
#Create a corleation matrix
corr_matrix = data.corr()
```

```python
# Heatmap of data
plt.figure(figsize=[16,6])
sns.heatmap(corr_matrix, annot=True , cmap ='coolwarm')
plt.title('Heatmap of the data')
plt.show()
```

```python
# Scatterplot of data
plt.figure(figsize=[16,6])
sns.scatterplot(x='Gender', y='Attendance', data=data)
plt.xlabel('Gender')
plt.ylabel('Attendance')
plt.title('Scatter Plot of Gender vs Attendance')
plt.show()
```
